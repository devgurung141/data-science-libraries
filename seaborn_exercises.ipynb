```python

def get_db_url(db, env_file=os.path.exists('env.py')):
    '''
    return a formatted string containing username, password, host, and 
    database for connecting to the mySQL server and the database indicated
    '''
    if env_file:
        username, password, host = (env.username, env.password, env.host)
        return f'mysql+pymysql://{username}:{password}@{host}/{db}'
    else:
        return 'An error'
```

```python
url = get_db_url('chipotle')
```

```python
query = 'SELECT * FROM orders'

chipotle_orders = pd.read_sql(query, url)

chipotle_orders .head()
```

```python
chipotle_orders.dtypes
```

```python
chipotle_orders.item_price = chipotle_orders.item_price.str.replace('$','').astype('float')
```

```python
chipotle_orders.dtypes
```

```python
# chipotle_orders.item_name.value_counts()
# popular = chipotle_orders.item_name.value_counts().head(4)
# chipotle_orders.item_name.value_counts().head(4).index
popular = chipotle_orders.item_name.value_counts().head(4)
popular
```

```python
revenue = chipotle_orders.groupby('item_name')['item_price'].agg('sum') # revenue produced by each
revenue
```
