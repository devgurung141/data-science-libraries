```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
n = np.array([x for x in range(1, 5)])
n
```

```python
x_series = pd.Series(n)
# x_series
```

```python
def factorial(n):
    return 1 if (n==1 or n==0) else n * factorial(n - 1);

y7 = x_series.apply(factorial)
y7
```

```python
y1 = 0*n + 1
# print('y1:',y1)
y2 = np.log(n)
# print('y2:',y2)
y3 = n
# print('y3:',y3)
y4 = n * np.log(n)
# print('y4:',y4)
y5 = n**2
# print('y5:',y5)
y6= 2**n
# print('y6:',y6)
y7 = x_series.apply(factorial)
# print('y7:',y7)
y8 = n^n
# print('y8:',y8)

plt.plot(n, y1, c='red')
plt.plot(n, y2, c='darkgreen')
plt.plot(n, y3, c='black')
plt.plot(n, y4, c='pink')
plt.plot(n, y5, c='orange')
plt.plot(n, y6, c='brown')
plt.plot(n, y7, c='cyan')
plt.plot(n, y8, c='yellow')

# plt.xlim(0,5)
# plt.ylim(0,100)
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(['$0n+1$', '$log(n)$', '$n$', '$n*log(n)$', '$n^2$', '$2^n$', '$n!$', '$n^n$'])
plt.text(1,10,'Dev', fontsize=8, c='maroon')

plt.show()



```
