```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(range(0, 11))
```

```python
y = x**2 - x + 2

plt.plot(x,y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.annotate('origin',xy =(0,0), xytext =(8,0),arrowprops =dict(arrowstyle="->") )
# plt.annotate('origin',xy =(0,0), xytext =(0,30),arrowprops ={}) 
plt.title('$y = x^2 -x +2$')
plt.grid(True, ls = '--')  #  display grid lines

plt.show()
```

```python
y = np.sqrt(x)

plt.plot(x, y, c='red')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = x$')
plt.grid(True, ls = '--')

plt.show()
```

```python
y = x**3

plt.plot(x,y, c='orange')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = x^3$')
plt.grid(True, ls = '--')

plt.show()
```

```python
y = 2**x

# to use smaller values
    # x3 = [x * 0.1 for x in range(0,,100)]
    # y3 = [2 ** n for n in x3]

plt.plot(x,y, c='green')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = 2^x$')
plt.grid(True, ls = '--')

plt.show()
```

```python
y = 1/(x+1)

plt.plot(x,y, c='darkblue')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y = 1/(x+1)$')
plt.grid(True, ls = '--')

plt.show()
```

```python
# plt.figure(figsize= (10,6))  # set size of graph

plt.subplot(221)
y = np.sqrt(x)
plt.plot(x, y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y = x$')
plt.plot(x,y, c='red')
plt.grid(True, ls = '--')

plt.subplot(222)
y = x**3
plt.plot(x, y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y = x^3$')
plt.plot(x,y, c='orange')
plt.grid(True, ls = '--')

plt.subplot(223)
y = 2**x
plt.plot(x,y, c='green')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y = 2^x$')
plt.grid(True, ls = '--')

plt.subplot(224)
y = 1/(x+1)
plt.plot(x,y, c='darkblue')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y = 1/(x+1)$')
plt.grid(True, ls = '--')

plt.suptitle('comparison between graphs')
plt.tight_layout()

plt.show()
```

```python
a = np.sqrt(x)
b = x**3
c = 2**x
d = 1/(x+1)

plt.plot(x, a, c='red')
# plt.plot(x, a, c='red', lable='$y=x$', alpha=1.0)
plt.plot(x, b, c='green')
plt.plot(x, c, c='orange')
plt.plot(x, d, c='darkblue')
plt.xticks(range(-5, 11, 1))
# plt.yticks(range(-5, 11, 1))
# plt.title(' Plot Graph')
plt.title('4 functions')
plt.grid(True, ls = '--')
plt.legend(['$y=x$','$y=x^3$', '$2^x$', '$y=1/(x+1)$'])

plt.show()
```
