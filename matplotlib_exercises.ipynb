```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt


```

```python
x = np.array(range(0, 11))
```

```python
y = x**2 - x + 2

plt.plot(x,y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y=x^2 -x +2$')



plt.show()
```

```python
y = np.sqrt(x)

plt.plot(x, y, c='red')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y=√x$')

plt.show()
```

```python
y = x**3

plt.plot(x,y, c='orange')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y=x^3$')

plt.show()
```

```python
y = 2**x

plt.plot(x,y, c='green')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y=2^x$')

plt.show()
```

```python
y = 1/(x+1)

plt.plot(x,y, c='darkblue')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y=1/(x+1)$')

plt.show()
```

```python
plt.subplot(221)
y = np.sqrt(x)
plt.plot(x, y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y=√x$')
plt.plot(x,y, c='red')

plt.subplot(222)
y = x**3
plt.plot(x, y)
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y=x^3$')
plt.plot(x,y, c='orange')

plt.subplot(223)
y = 2**x
plt.plot(x,y, c='green')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y=2^x$')

plt.subplot(224)
y = 1/(x+1)
plt.plot(x,y, c='darkblue')
plt.xticks(range(-5,11, 1))
# plt.yticks(range(-5,11, 1))
# plt.xlabel('$x$')
# plt.ylabel('$y$')
plt.title('$y=1/(x+1)$')

plt.suptitle('comparison between graphs')
plt.show()

```

```python
a = np.sqrt(x)
b = x**3
c = 2**x
d = 1/(x+1)


plt.plot(x, a, c='red')
plt.plot(x, b, c='green')
plt.plot(x, c, c='orange')
plt.plot(x, d, c='darkblue')
plt.xticks(range(-5, 11, 1))
# plt.yticks(range(-5, 11, 1))
plt.title('Plot Graph')

plt.legend(['$y=√x$','$y=x^3$', '$2^x$', '$y=1/(x+1)$'])
plt.show()

```
